### Scraping the Data

<b> Question 2</b> Using the contents and `beautiful soup` load the data from the `By market capitalization` table into a `pandas` dataframe. The dataframe should have the country `Name` and `Market Cap (US$ Billion)` as column names.  Display the first five rows using head. 


```python
#The Beautiful Soup library allowd us to pull data out of HTML and XMl files.

#Replace the dots below

#Create a BeautifulSoup object using the variable html_data
soup=BeautifulSoup(html_data, "html5lib")

#Scrape data from HTML tables, 
#But first to make things easier, check the website to see how many rows and columns are in the table

#Now find a html table in the webpage, which is represented by the tage <table>
tables = soup.find_all('table')
len(tables)      #How many tables we found by checking the length of the tables list

#We are looking for the By market capitalization table
table_index = 2
for index, tabl in enumerate(tables):
    if("By market capitalization" in str(tabl)):
        table_index = index
print(table_index)

print(tables[table_index].prettify())

```

Using BeautifulSoup parse the contents of the webpage.


Load the data from the `By market capitalization` table into a pandas dataframe. The dataframe should have the country `Name` and `Market Cap (US$ Billion)` as column names. Using the empty dataframe `data` and the given loop extract the necessary data from each row and append it to the empty dataframe.


```python
#Loading the data from the HTML table into a panda datafram
data = pd.DataFrame(columns=["Name", "Market Cap (US$ Billion)"])

#Get all rows from the table
for row in soup.find_all('tbody')[3].find_all('tr'): #in html table, row is rep by <tr>
    print("Row here is:" )
    print(row)
    #Get all columns (rep by <td>) in each row
    col = row.find_all('td')  
    if (col!=[]):
        country = col[1].find('a', href= True).get('title') #Get the country name by extracting the title from href
        cap = col[2].text.strip()   #store the value in col 2 as cap and remove "\n"
        data = data.append({"Name":country, "Market Cap (US$ Billion)":cap}, ignore_index=True)
        
    #Write your code here
data
```
